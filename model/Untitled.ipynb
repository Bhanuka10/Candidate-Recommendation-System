```python
import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

def preprocess_text(text):
    text = text.lower()
    text = re.sub(r'\[|\]|\'', ' ', text)     # remove list symbols
    text = re.sub(r'[^a-zA-Z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    tokens = text.split()
    tokens = [w for w in tokens if w not in ENGLISH_STOP_WORDS]
    return " ".join(tokens)
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def remove_stopwords_lemmatize(text):
    words = text.split()  # split into words
    cleaned = []

    for w in words:
        if w not in stop_words:                 # remove stopwords
            cleaned.append(lemmatizer.lemmatize(w))  # lemmatize

    return " ".join(cleaned)
```

```python
df["cv_text"] = df["cv_text"].apply(remove_stopwords_lemmatize)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
def stem_text(text):
    words = text.split()
    stemmed_words = [stemmer.stem(w) for w in words]
    return " ".join(stemmed_words)
df["cv_text"] = df["cv_text"].apply(stem_text)
```

```python
df["cv_text"].head()
```

```python
df["cv_clean"] = df["cv_text"].apply(preprocess_text)
df["cv_clean"].head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(
    ngram_range=(1,2),
    max_df=0.85,
    min_df=2
)

cv_vectors = vectorizer.fit_transform(df["cv_clean"])
```
