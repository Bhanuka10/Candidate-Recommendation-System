```python
# Define preprocessing function
def preprocess_text(text):
    if pd.isna(text):
        return ''
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Remove special characters (keep letters, numbers, spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    
    # Remove extra whitespaces
    text = ' '.join(text.split())
    
    return text

# Apply to all text columns
data['Job Title'] = data['Job Title'].apply(preprocess_text)
data['Skills'] = data['Skills'].apply(preprocess_text)
data['Education Requirements'] = data['Education Requirements'].apply(preprocess_text)

# Check results
print(data.head())

# Save cleaned data
data.to_csv('jdfinal_cleaned.csv', index=False)
```

```python
data = pd.read_csv('jdfinal_cleaned.csv')
```

```python
data['combined_text'] = (
    data['Skills'] + ' ' + 
    data['Education Requirements']
)
```

```python
tfidf_vectorizer = TfidfVectorizer(
    max_features=200,        # Keep top 200 most important features
    min_df=2,                # Word must appear in at least 2 documents
    max_df=0.8,              # Ignore words that appear in more than 80% of documents
    ngram_range=(1, 2),      # Use single words and 2-word combinations
    lowercase=True,          # Convert to lowercase (already done, but safe)
    stop_words='english'     # Remove common English stop words
)
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(data['combined_text'])
```

```python
# Get feature names (words/terms)
feature_names = tfidf_vectorizer.get_feature_names_out()

# Convert sparse matrix to DataFrame
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=feature_names
)
```

```python
# Add the 'Required Years of Experience' column back
final_data = pd.concat([
    data[['Required Years of Experience']].reset_index(drop=True),
    tfidf_df
], axis=1)

print("âœ… Added original columns")
print(f"\nFinal Dataset Shape: {final_data.shape}")
print(final_data.head())
```

```python
# Calculate feature importance (sum of TF-IDF scores across all documents)
feature_importance = pd.DataFrame({
    'feature': feature_names,
    'importance': tfidf_matrix.sum(axis=0).A1
}).sort_values('importance', ascending=False)

print("\nðŸ“Š Top 20 Most Important Features:")
print(feature_importance.head(20))
```

```python
# Save the TF-IDF vectorized data
final_data.to_csv('jdfinal_tfidf_vectorized.csv', index=False)

print("\nðŸ’¾ Vectorized data saved as 'jdfinal_tfidf_vectorized.csv'")
print(f"âœ… Total columns: {final_data.shape[1]}")
print(f"âœ… Total rows: {final_data.shape[0]}")
```
